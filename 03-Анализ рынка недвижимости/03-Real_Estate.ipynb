Высота потолков - почти в половине случаев этот параметр не указан. Придётся заменять эти пропуски в дальнейшем, или оставить пустое значение. Удалять строки нельзя.

Количество балконов - скорее всего, если в столбце нет значения, то балконов нет и можно заменить на ноль.

Название населенного пункта - не особо важно для анализа, как называется пункт, важнее расстояние до центра. Можно заменить пустые значения на "Другое", например.

Расстояние до ближайшего аэропорта, парка, водоема  - врятли сильно влияют на стоимость квартиры, а пропусков много. Нецелесообразно тратить много времени на выявление параметров, с помощью которых можно заменить пустые значения в этих колонках. 

Сколько дней было размещено объявление - если цена адекватная, а с квартирой всё в порядке, то объявление должно висеть примерно столько же, сколько и другие аналогичные. Но возможно, квартира еще не была продана, или у нее есть какие-то неизвестные нам особенности, которые мешают быстро осуществить продажу. Поэтому не совсем корректно было бы заменять пустые значения на средние в зависимости от какого-либо параметра. Но и удалять строки с пустыми значениями не стоит - таких строк достаточно много, они могут содержать другие важные данные. Поэтому пустые значения оставим как есть.

## Предобработка данных

Сначала разберемся с пустыми значениями. 

```python
data.isna().sum() #находим, в каких столбцах есть пропущенные значения
```

```python
data.isna().sum() / data['last_price'].count() #посчитаем долю пропущенных значений в каждом столбце
```

Видно, что многие колонки содержат большую долю пропусков. Небольшая доля только в столбцах с наименованием локации и общей этажности зданий. Название локации мы никак не вычислим, а вот пропуски в количестве этажей можем попробовать заполнить.

```python
import seaborn as sns #способ визуального представления пропусков от ревьюера предыдущего проекта с целью выявления, не в одних и тех же столбцах эти пропуски.
sns.heatmap(data.isna(), yticklabels=False, cbar=False, cmap="YlGnBu");
```

Заполним пропуски в столбце с количеством балконов, потому что их отсутствие, скорее всего, означает, что балконов ноль.

```python
data['balcony'] = data['balcony'].fillna(value=0)
```

```python
data['balcony'].isna().sum() #проверяем, что пропуски в этом столбце исчезли
```

Высота потолков - заменим на медианное значение, т.к. пропусков много. Видимо, это поле было необязательным для заполнения.


```python
#data['ceiling_height'] = data['ceiling_height'].fillna(value=2.65)
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
```

```python
data['ceiling_height'].isna().sum() #проверяем, что пропуски в этом столбце исчезли
```

Количество этажей - посмотрим, какие там уникальные значения.

```python
data['floors_total'].sort_values().unique()
```

Всё, что выше 35 этажей, для Санкт-Петербурга странно.
Посмотрим, на каких этажах располагаются квартиры в таких домах.

```python
data.query('floors_total > 35')
```

Строк с очень высокими домами всего 6. Для анализа нам важно, первый/последний этаж или нет, поэтому не будем заменять высокие этажи, даже если там ошибка. По этим строкам видно, что этаж не последний.
Теперь посмотрим на строки с пропущенными значениями в этом столбце.

```python
data[data['floors_total'].isna()].head(15)
```

Мы не можем заменить общее количество этажей на среднее значение или медиану, т.к. этаж квартиры может оказаться выше, чем среднее. Поэтому целесообразно заменить на 35 - наиболее правдоподобное максимальное число этажей.

```python
data['floors_total'] = data['floors_total'].fillna(value=35) #заменяем пропуски на 35
```

```python
data['floors_total'].isna().sum() #проверяем, что пропуски в этом столбце исчезли
```

Теперь посмотрим на столбец is_apartment - является ли объект аппартаментами.

```python
data['is_apartment'].unique()
```

В большинстве случаев это значение пустое. Скорее всего, если у типа жилья не указали True, значит, что это не аппартаменты, то есть False. Заменим пустые значения на False.

```python
data['is_apartment'] = data['is_apartment'].fillna(value=False)
```

```python
data['is_apartment'].unique() #проверяем, что пустые значения исчезли
```

```python
data['is_apartment'] = data['is_apartment'].astype('bool')  #заменим тип на bool, т.к. значения True и False
```

Расстояние до центра важно, если квартира в Санкт-Петербурге. Посмотрим, сколько таких пропусков.

```python
data[((data['locality_name'] == 'Санкт-Петербург') 
               & (data['cityCenters_nearest'].isna()))]
```

Всего 61 пропуск. Удалим такие строки, чтобы не мешали проводить анализ.

```python
data = data[(data['locality_name'] != 'Санкт-Петербург') 
            | ((data['locality_name'] == 'Санкт-Петербург') 
               & (~data['cityCenters_nearest'].isna()))].reset_index(drop=True)
```

```python
data['cityCenters_nearest'].isna().sum() #проверим, сколько пустых значений осталось в столбце с расстоянием до центра
```

Остальные отсутствующие значения оставим пустыми, т.к. в пригороде расстояние до центра не должно сильно влиять на цену.

Есть пропуски в названии населенных пунктов. Заменим их на "Другой", т.к. по данным в таблице не определить, к какому именно пункту относится та или иная квартира.
Возможно, эти данные оказались пустыми из-за ошибки выгрузки, ведь название населенного пункта - слишком важный параметр при продаже, чтобы пропускать объявление на сайт без его заполнения. Но для нашего анализа название не так важно.

```python
data['locality_name'] = data['locality_name'].fillna(value='другой') #заменим отсутствующие населенные пункты на "другой"
```

```python
data['locality_name'].isna().sum() #определим количество пропусков в наименовании населенного пункта
```

Чтобы заполнить пустые значения в колонке с площадью кухни, найдём среднее соотношение площади кухни к общей площади, а затем умножим общую площадь на среднее значение соотношения в строках, где площадь кухни отсутствует.
